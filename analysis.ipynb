```python
def duration_to_minutes(val):
    if pd.isnull(val):
        return None
    val_str = str(val)
    if '.' in val_str:
        hours, minutes = val_str.split('.')
        return int(hours) * 60 + int(minutes)
    else:
        return int(val_str) * 60  # No decimal part

df['duration_minutes'] = df['duration'].apply(duration_to_minutes)
```

```python

source = input("Enter source city: ").lower()
destination = input("Enter destination city: ").lower()

filtered_df = df[
    (df['source_city'].str.lower() == source) &
    (df['destination_city'].str.lower() == destination)
]


print(f"\nRoute: {source.title()}  {destination.title()}")
print(f"Total flights found: {len(filtered_df)}")

if not filtered_df.empty:
    cheapest = filtered_df.loc[filtered_df['price'].idxmin()]
    avg_price = filtered_df['price'].mean()

    print(f" Cheapest ticket: {cheapest['airline']} | {cheapest['class']} class | {cheapest['price']}")
    print(f" Average price: {avg_price:.2f}")


    plt.figure(figsize=(12, 6))
    sns.barplot(data=filtered_df, x='airline', y='price', hue='class', estimator='mean')
    plt.title(f"Average Ticket Prices for {source.title()}  {destination.title()}")
    plt.ylabel('Average Price')
    plt.xlabel('Airline')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print(" No flights found for this route.")
```
